```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
col_names = ['Age', 'Gender', 'Payment Method', 'Future Customer']
# load dataset
deals = pd.read_csv("/Users/anos/Documents/Works/Training/dataset/Deals.csv")
```

```python
deals
```

```python
#split dataset in features and target variable
feature_cols = ['Age', 'Gender', 'Payment Method']
X = deals[feature_cols] # Features
y = deals['Future Customer'] # Target variable
```

```python
#from sklearn.preprocessing import OrdinalEncoder
#from sklearn.preprocessing import OneHotEncoder
#transformer = OneHotEncoder().fit(X)
#X = transformer.transform(X)
```

```python
X = pd.get_dummies(X)
```

```python
X.columns
```

```python
feature_cols = X.columns
```

```python
#y[(y=="yes")] = 1
#y[(y=="no")] = 0
#y=y.astype('int')
```

```python
y.hist()
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=77) # 70% training and 30% test
```

```python
y_train
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=4)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from six import StringIO
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('deals.png')
Image(graph.create_png())
```
